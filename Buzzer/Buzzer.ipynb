## Buzzer 
This is a Code that can Run your Buzzer Module on your Raspberry Pi. I have Personally Used it and it works just fine with my Raspberry pi and Buzzer Module. 

```python
GPIO.setwarnings(False)
GPIO.setmode(GPIO.BOARD)
```

```python
buzzer=8
GPIO.setup(8,GPIO.OUT)
#Run forever loop
```

```python
while True:
    GPIO.output(8,GPIO.HIGH)
    print ("Beep")
    sleep(1) # Delay in seconds
    GPIO.output(8,GPIO.LOW)
    print ("No Beep")
    sleep(0.5)
```

This should be able to Beep Your Buzzer Up.
